```python
import pandas as pd

data = pd.read_csv('ggs_product_data.csv')
data.head()
```

```python
product_list = []

for i in range(len(data)):
    # skip if any required field is null
    if data.loc[i, [
        'Group', 'Crop Name', 'English', 'Hindi', 
        'Marathi', 'Variety Name'
    ]].isnull().any():
        continue

    data_dict = {
        'Group_Name': data['Group'][i],
        'Name': data['Crop Name'][i],
        'Description_English': data['English'][i],
        'Description_Hindi': data['Hindi'][i],
        'Description_Marathi': data['Marathi'][i],
        'Variety_Name': data['Variety Name'][i],
    }
    product_list.append(data_dict)
```

```python
import json
def put_product(product_id,product_data,lang):
    print(product_data)
    product ={
        'Group_Name': product_data['Group_Name'],
        'Name': product_data['Name'],
        'Variety_Name': product_data['Variety_Name'],
    }
    if lang == 'hi':
        product['Description'] = product_data['Description_Hindi']
    elif lang == 'mr':
        product['Description'] = product_data['Description_Marathi']
    url = f"http://localhost:1337/api/products/{product_id}?locale={lang}"
    

    payload = json.dumps({
    "data": product
    })
    headers = {
    'Content-Type': 'application/json'
    }
    headers["Authorization"] = f"Bearer {api_key}"


    response = requests.request("PUT", url, headers=headers, data=payload)

    print(response.text)
```

```python
import requests
import json

product_data = {
    'Name': 'test',
    'Description': 'test',
    'Variety_Name': 'test',
}
def create_product(product_data,lang):
  print(product_data)
  url = f"http://localhost:1337/api/products?locale=en"
  product = {
    'Group_Name': product_data['Group_Name'],
    'Name': product_data['Name'],
    'Variety_Name': product_data['Variety_Name'],
    'Description': product_data['Description_English'],
  }
  payload = json.dumps({
    "data": product
    
  })
  headers = {
    'Content-Type': 'application/json'
  }
  headers["Authorization"] = f"Bearer {api_key}"

  response = requests.request("POST", url, headers=headers, data=payload)
  response.json()
  res = response.json()
  print(res)
  pid = res['data']['documentId']
  for lange in ['hi','mr']:
    put_product(pid,product_data,lange)
```
